Research Question # 1 : Does having close friends correlate to how optimistic you are about your future?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Manage data for variable 1
df['close_friends'] = df['H4WS4']
```

```python
df['close_friends'].value_counts()
```

```python
df['close_friends'] = df['close_friends'].loc[ df['close_friends'] !=' ']
df['close_friends'] = df['close_friends'].loc[ df['close_friends'] !='98']
df['close_friends'] = df['close_friends'].loc[ df['close_friends'] !='96']
df['close_friends'] = df['close_friends'].loc[ df['close_friends'] !='95']
df['close_friends'].value_counts()
```

```python
#Manage data for variable 1
```

```python
df['siblings'] = df['H4WS1']
```

```python
df['siblings'] = df['siblings'].loc[df['siblings'] != ' ']
```

```python
df['siblings'] =pd.to_numeric(df['siblings'],errors = 'coerce')
```

```python
df['siblings'] = df['siblings'].loc[ df['siblings'] !='6']
df['siblings'] = df['siblings'].loc[ df['siblings'] !=' ']
df['siblings'] = df['siblings'].loc[ df['siblings'] < 5]
df['siblings'].value_counts()
```

```python
df['siblings'].replace(['0','1','2','3','4'] , ['NONE','1 ', '2', '3','4',], inplace = True)
df['siblings'].value_counts()
```

```python
#Graphs and tables or statistics
```

```python
explode = (0,1,2,3,4)
plt.title("How many siblings do you have?")
df['siblings'].value_counts().plot(kind ='pie', autopct='%1.1f%%', colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown'])
plt.legend(loc ="upper right")
```

```python
plt.xlabel("Amount of Siblings")
plt.ylabel("Percent")
plt.title("How many siblings do you have?")
df['siblings'].value_counts(normalize = True).plot(kind ='bar', color = 'blue', width = 0.6, alpha = 0.5)
```

```python
plt.xlabel("Amount of Siblings")
plt.ylabel("Percent")
plt.title("How many siblings do you have?")
sns.histplot(x='siblings', data=df  , bins= 5, color='green')
```

```python
plt.xlabel("Amount of Siblings")
plt.ylabel("Percent")
plt.title("How many siblings do you have?")
sns.boxplot(x = 'siblings', data = df)
```
