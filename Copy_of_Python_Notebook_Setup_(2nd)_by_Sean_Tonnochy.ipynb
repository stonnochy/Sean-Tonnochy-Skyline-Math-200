Research Question # 2 : Is there an association between having siblings and how you react to things that already happened to you?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

Codebook 1

```python
#Manage data for variable 1
```

```python
df['dont_worry'] = df['H4PE30']
```

```python
df['dont_worry'].value_counts()
```

```python
df['dont_worry'] = df['dont_worry'].loc[ df['dont_worry'] !=' ']
df['dont_worry'] = df['dont_worry'].loc[ df['dont_worry'] !='6']
df['dont_worry'] = df['dont_worry'].loc[ df['dont_worry'] !='8']
df['dont_worry'].value_counts()
```

```python
#Basic Statistical Measure
df['dont_worry'].replace(['1','2','3','4','5'] , ['AGREE', 'AGREE', 'NEITHER','DISAGREE','DISAGREE'], inplace = True)
df['dont_worry'].value_counts()
```

```python
#Graphs and tables or statistics
```

```python
#Pie chart
explode = (0.0,0.0,0.0,0.4)
plt.title("I don't worry about things that have already happened.")
df['dont_worry'].value_counts().plot(kind ='pie', autopct='%1.1f%%', colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown','pink'])
plt.legend(loc ="upper right")
```

```python
#Bar chart
df['dont_worry'].value_counts().plot(kind ='bar')
```

Codebook 2:

```python
#Manage data for variable 1
```

```python
df['siblings'] = df['H4WS1']
```

```python
df['siblings'] = df['siblings'].loc[df['siblings'] != ' ']
```

```python
df['siblings'] =pd.to_numeric(df['siblings'],errors = 'coerce')
```

```python
df['siblings'] = df['siblings'].loc[ df['siblings'] !='6']
df['siblings'] = df['siblings'].loc[ df['siblings'] !=' ']
df['siblings'] = df['siblings'].loc[ df['siblings'] < 5]
df['siblings'].value_counts()
```

```python
df['siblings'].replace(['0','1','2','3','4'] , ['NONE','1 ', '2', '3','4',], inplace = True)
df['siblings'].value_counts()
```

```python
#Graphs and tables or statistics
```

```python
explode = (0,1,2,3,4)
plt.title("How many siblings do you have?")
df['siblings'].value_counts().plot(kind ='pie', autopct='%1.1f%%', colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown','purple'])
plt.legend(loc ="upper right")
```

```python
plt.xlabel("Amount of Siblings")
plt.ylabel("Percent")
plt.title("How many siblings do you have?")
df['siblings'].value_counts(normalize = True).plot(kind ='bar', color = 'blue', width = 0.6, alpha = 0.5)
```

```python
plt.xlabel("Amount of Siblings")
plt.ylabel("Percent")
plt.title("How many siblings do you have?")
sns.histplot(x='siblings', data=df  , bins= 5, color='green')
```

```python
plt.xlabel("Amount of Siblings")
plt.ylabel("Percent")
plt.title("How many siblings do you have?")
sns.boxplot(x = 'siblings', data = df)
```
